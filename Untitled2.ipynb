```python
import pandas as pd
from io import StringIO

# Sample CSV data
csv_data = """
CustomerID,Gender,Age,Annual Income (k$),Spending Score (1-100)
1,Male,19,15,39
2,Male,21,15,81
3,Female,20,16,6
4,Female,23,16,77
5,Female,31,17,40
6,Female,22,17,76
7,Female,35,18,6
8,Female,23,18,94
9,Male,64,19,3
10,Female,30,19,72
"""

# Read CSV into DataFrame
df = pd.read_csv(StringIO(csv_data))
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check the first few rows
print("🔹 First 5 rows of the data:")
display(df.head())

# Check data types and missing values
print("🔹 Data Info:")
df.info()

# Summary statistics
print("🔹 Summary Statistics:")
display(df.describe())

# Count of Gender
sns.countplot(data=df, x='Gender')
plt.title("Gender Distribution")
plt.show()
# Scatter Plot: Income vs Spending Score
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', s=100)
plt.title("Customer Spending vs Income")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Select features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Apply K-Means with 3 clusters (you can try more later)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='Set1', s=100)
plt.title("Customer Segments")
plt.show()
```

```python
# 1. Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# 2. Spending Score by Gender
plt.figure(figsize=(8,5))
sns.boxplot(x='Gender', y='Spending Score (1-100)', data=df)
plt.title("Spending Score by Gender")
plt.show()

# 3. Age vs Spending Score with Clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)', hue='Cluster', palette='Set2', s=100)
plt.title("Clusters based on Age and Spending")
plt.show()
```
